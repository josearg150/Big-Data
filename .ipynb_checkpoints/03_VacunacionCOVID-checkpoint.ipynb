```python
#Importamos las librerias
import pandas as pd
import numpy as np
#inmport statsmodels as ss #estadisiticas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#leemos el dataset
data = pd.read_csv('D:\\7 Semestre\\Big Data\\Jupiter\\country_vaccinations.csv')
```

```python
data
```

```python
data.dtypes
```

```python
#preprocesamiento de informacion
#ejemplo de reemplazo y cambio de tipo de dato
data["date"] = data["date"].str.replace('/','-')
```

```python
data
```

```python
#Limpieza de la informacion
data.isnull().sum() #suma si son nulos
```

```python
#Estadisticas : muestra los porcentajes de la informacion faltante por columna
data.isnull().sum()/data.shape[0] #shape : (rows,cols)
```

```python
#eliminar los Nan que existen
datosLimpios = data.dropna()
```

```python
datosLimpios
```

```python
datosLimpios.shape
```

```python
#revisamos los datos que quedan
datosLimpios.isnull().sum()
```

```python
conjunto_datosLimpios = datosLimpios[["country", "date", "total_vaccinations"]]
conjunto_datosLimpios
```

```python
#subconjunto mexico
mx = conjunto_datosLimpios.loc[conjunto_datosLimpios["country"] == "Mexico"]
mx
```

```python
#Conversion de float a int
mx["total_vaccinations"].astype(int)
```

```python
#Representacion visual de los datos
#ejes
x = mx["date"]
y = mx["total_vaccinations"]
#Creando el plot
plt.figure(figsize = (20,6))
plt.plot(x,y)
ax = plt.gca() #activa la proyeccion del plot para configuracion
ax.get_yaxis().get_major_formatter().set_scientific(False) #para que el eje
plt.xticks(np.arange(0, len(x), step = 20), rotation = 45)
plt.ylim(0, max(y))
plt.title("Total de vacunaciones en Mexico")
plt.xlabel("Fecha")
plt.ylabel("Vacunaciones")

plt.show()
```

```python
#Normalizando la informacion
#Agregamos columnas y las transformamos (las variables)
#A datos logaritimicos con base 10
# corr + fuerte
ccDatosLimpios['total_vacc'] = np.log10(ccDatosLimpios['total_vaccinations'])
ccDatosLimpios['people_vacc'] = np.log10(ccDatosLimpios['people_vaccinated'])
ccDatosLimpios['people_fully_vacc'] = np.log10(ccDatosLimpios['people_fully_vaccinated'])
ccDatosLimpios['daily_vacc'] = np.log10(ccDatosLimpios['daily_vaccinations'])
```

```python
#Falta crear otro dataset para que funcione
fig, axes = plt.subplots(ncols=len(ccDatosLimpios.columns), figsize=(30,5))
for ax, col in zip(axes, ccDatosLimpios.columns):
    sns.distplot(ccDatosLimpios[col], ax = ax) 
    plt.tight_layout()
plt.show()
```

```python
cc = ccDatosLimpios.drop(['country', 'iso_code', 'date', 'vaccines', 'source_name', 'source_website'], axis = 1)
cc = cc.replace([np.inf, -np.inf],0)
```

```python
cc.columns
```

```python
#Falta crear otro dataset para que funcione
fig, axes = plt.subplots(ncols=len(cc.columns), figsize=(30,5))
for ax, col in zip(axes, cc.columns):
    sns.distplot(cc[col], ax = ax) #Hay que eliminar datos str
    plt.tight_layout()
plt.show()
```

```python
#Creamos un mapa de calor para ver las correlaciones entre variables
corr = cc.corr()
fig = plt.figure(figsize = (12,10))
ax = fig.add_subplot(111)
cax = ax.matshow(corr, cmap='coolwarm', vmin=1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(cc.columns), 1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.show()
```

```python
tablaCorr=cc.corr(method='pearson') #relacion lineal entre dos variables
tablaCorr
```

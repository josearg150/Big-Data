```python
#analizando la informacion
ruta = "d:\\7 Semestre\\Big Data\\Jupiter\\"
df = pd.read_csv(ruta+"titanic.csv")
print(df.head(5))
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.tail())#ultimos 5
```

```python
#fase 1 limpiza de datos: datos vacios, duplicados, datos y formato equivocado
#revisar la informacion df 
#datos vacios-eliminar renglon 
df_notNan = df.dropna() #Sobre el df original df.dropna(inplace=true)
#print(df_notNan.to_string())
print(df_notNan.tail())
```

```python
print(df_notNan.info())
```

```python
#rellena valores vacios 
#fillna() reemplaza la celda vacia con un valor 
df.fillna(-1, inplace = True)
```

```python
df
```

```python
#seleccionando columnas 
df["age"].fillna(-1, inplace = True)
```

```python
df
```

```python
#reemplazando los datos o celdas vacias con la media 
#media promedio de una serie de valores 
m = df["age"].mean()
df["age"].fillna(m, inplace = True)
```

```python
df
```

```python
#mediana : valor medio de una serie de valores ordenados
m = df["age"].median()
df["age"].fillna(m, inplace = True)
```

```python
#moda : valor mas popular de una serie de valores 
m = df["age"].mode()[0]
df["age"].fillna(m, inplace = True)
```

```python
#datos equivocados
#titanic: clases: 1 2 3 < valores permitidos. valore equivocado 0 4 5
#para modificar un registro: registro 1
#df.loc[1, "pclass"] = 3
#para modificar toda la columna 
for i in df.index:
    if df.loc[i, "pclass"] > 3 or df.loc[i, "pclass"] < 1:
        if df.loc[i, "class"] == "Third":
            df.loc[i, "pclass"] = 3
        elif df.loc[i, "class"] == "Second":
            df.loc[i, "pclass"] = 2
        elif df.loc[i, "class"] == "First":
            df.loc[i, "pclass"] = 1
```

```python
df
```

```python
#para borrar datos equivocados y eliminar renglon completo 
for i in df.index:
    if df.loc[i, "age"] < 5.0:
        print(i)
        df.drop(i, inplace = True)
```

```python
df
```

```python
#para eliminar renglones duplicados
print(df.duplicated())
```

```python
#para eliminar renglones duplicados
df.drop_duplicates(inplace = True)
```

```python
print(df)
```

```python
#fase 2. encontrar relaciones en la informacion (coorelacionar)
#metodo corr(): calcula la relacion entre columnas
#valores -1.0 a 1.0
#relacion perfecta : 1.000, buena relacion: 0.90000
#mala relacion: 0.000291 <- significa 
df.corr()
```

Sistema diseado para procesar datos de manera distribuida sobre clusters. Spark puede procesar cantidades de datos en el orden de terabytes incluso petabytes.

El concepto de Spark es imaginar un cluster como una memoria gigante, la memoria resultante de combinar las memorias de todos los clusters. Se prioriza el uso de memoria y consigue ser muy rpido el procesamiento de la inforamcin, mayor que si utilizara MapReduce (Google).

Spark utiliza MapReduce para algunas ateas de clasificacin mediante regresin. Apache Spark esta implementado en Scala que es ejecutado en la mquina virtual de Java. Adems de Spark ofrece interfaces de programacin para Java, Python y R.

Caracteristicas: velocidad de procesamiento, soporte multilenguaje, anlisis avanzado. Se puede desarrollar en tres maneras:
1:)solo : standalone. Utiliza como base HDFS y encima se encuentra spark.
2:)Hadoop (Yarn, adminsitrador de recursos) : Base HDFS -> Yarn/Mesos(kernel administrador del cluster) -> Spark
3:)Spark con MapReduce (Spark In MapReduce, SIMR) : Base HDFS -> MapReduce y dentro se encuentra Spark

Componentes de spark
Cuando se ejecuta Spark en un ambiente distribuida se distinguen dos tipos de procesos: driver y executor. Un proceso driver conectado a 3 procesos excutor localizados en dos nodos del cluster:

driver: proceso principal. Este proceso tiene un objeto SparkContext que te permite conectar con el gestor del cluster y reservar procesos executor en los distintos nodos del cluster. Cada uno de los nodos del cluster (tambien se conocen como worker) pordr ejecutar uno o varios procesos executor que almaenar fragmentos de los datos del programa y realizar operaciones sobre ellos. Durante la ejecucin ir enviando peticiones a los distintos procesos executor que pueden contactar enre ellos para realizar tareas y comunicar con el proceso driver para devolver resultados.

RDD (Resilient Distributed Datasets)
Tipo de datos bsicos. Estos datos almacenan informacin de manera distribuida entre todos los equipos del cluster. Durante la ejecucin de un programa Sparck se construyen varios RDDs que se dividen en distintos fragmentos y son almacenados en la memoria de los equipos del cluster.

Caracteristicas
Estn formados por un conjunto de registros, tambin llamados elementos, todos del mismo tipo. Por ejemplo, si cargamos un RDD a partir de un archivo plano se crear un RDD de cadenas de texto, una por cada linea del archivo.

En Scala o Java, al declarar un RDD se debe definir el tipo de los registros:

Lenguajes estticos : RDD[String] en Scala y JavaRDD en Java.
Lenguajes dinmicos : Python se pueden mezclar los tipos de datos.
RDDs han sido diseados desde el inicio para ser distribuidos: Los registros que lo componen se repartirn entre los clster.

Para realizar esta distribucin: los RDDs se dividen en particiones. Cada particin se almacena nciamente en un proceso executor dentro de un nodo del clster, aunque un proceso executor puede albergar distintas particiones de distintos RDDs. El nmero de particiones en las que dividir un RDD se puede configurar e incluso cambiar a lo largo de la ejecucin, por default es el nmero de ncleos de procesamiento disponibles en el clster.

Para decidir qu registros forman parte de cada particin, Spark utiliza particionadores, que son funciones que toman un registro y devuelven el nmero de la particin a la que pertenecen. Estos particionadores se puede configurar si se desea mejorar el rendimiento o dejar el default.

Los RDDs son inmutables: no se puede modificar ni actualizar. Una vez creado, as permanece hasta que se termina la ejecucin del programa.

Ejemplo
Cmo se podra particionar un RDD de 13 parejas (int, str) sobre 3 procesos executor utilizando el rango de valores del primer elemento de la pareja.



Los RDDs son inmutables: no se puede modificar ni actualizar. Una vez creado, as permanece hasta que termina la ejecucin del programa.

RDDs admiten dos tipos de operaciones:
    transformaciones y acciones
pero ninguna de ellas modifica el RDD. 
*Transformaciones:
    son operaciones que toman un RDD de partida y crean un nuevo
    RDD, dejando el original intacto. 
    Ejemplos: son aplicar una
    funcin a todos los registros del RDD
    (por ejemplo, sumar una cierta cantidad), 
    filtrar nicamente aquellos registros que cumplan una cierta condicin u ordenarlos
    mediante algn campo.
*Acciones:
    son operaciones que realizan algn cmputo sobre
    el RDD y devuelven un valor, dejando tambin el 
    RDD original inalterado. 
    Ejemplos : sumar todos los elementos almacenados 
    en un RDD de nmeros, generando un valor final, o
    "vaciar" un RDD a un archivo de texto. 
    
    
Concepto importante del RDD: Resilencia
    Tienen la capacidad de recuperar su estado inicial
    cuando existe algn problema.
    Esto se debe a que los RDDs son particionados y cada 
    particin ha sido almacenada en un proceso executor. 
    Por lo que apartir de su estado inicial, repite las
    transformacines que tiene programadas. De esta manera,
    regenera las particiones perdidas.

Una de las acciones ms tiles sobre RDDs es:reduce permite recorrer todos los valores de un RDD y calcular un valor en relacin con ellos. Reduce un RDD a un nico valor. Dado un RDD con elementos de tipo T, la funcin de reduccin es una funcin binaria que acepta dos elementos de tipo T y devuelve un valor del mismo tipo T, es decir, tiene tipo T x T -> T. reduce(f) es un mtodo que aplica la funcin de reduccin f a los 2 primeros elementos, luego aplica de nuevo la funcin f al valor result
ante y al tercer elemento, y as sucesivamente hasta que procesa el ltimo elemento y produce el valor final. Ejemplo: Se muestra el proceso para calcular la suma de un RDD de 5 elementos enteros. 



```python
#Para crear un RDD con los nmeros del 1 al 5 y calcular su suma 
#definiriamos una funcin de reduccin
#add e invocaramos a reduce como sigue:
def add(x, y):
    return x + y

#llamado a reduce
r = sc.parallelize(range(1, 6))
r.reduce(add)
```

```python
#Utilizando una funcin annima o lambda
r = sc.parallelize(range(1,6))
r.reduce(lambda x, y: x + y)
```

```python
#Mtodo takeOrdered, take()
r = sc.parallelize(range(1000))
r.takeOrdered(10, lambda x : -x)
```

```python
r = sc.parallelize(range(1000))
r.takeSample(False, 10)
```

```python
def multiply_positive(x, y):
    if x > 0 and y > 0:
        return x*y
    elif x > 0:
        return x
    elif y > 0:
        return y
    else:
        return 1
    
r = sc.parallelize([-1, 2, 1, -5, 8])
r.reduce(multiply_positive)
```

reduce  se requiere que la funcin sea conmutativa y asociativa para los dos valores
cualquiera x,y, f(x,y) = f(y,x) y para una triada : x,y,z, f(x,f(y,z) ) = f(f(x,y),z).
Ejemplos de funciones conmutativas y asociativas son la suma, la multiplicacin, el mnimo o el mximo.
La resta o la divisin no son conmutativas ni asociativas.(no se pueden reducir)
***
* MapReduce requiere de acciones como la replicacin, serializacin y E/S de almacenamiento para poder ejecutar 
* las operaciones solicitadas.
* stas ltmas son las que ms requiere el sistemade archivos de Hadoop. 
* Cualquier operacin de MapReduce requiere del almacenamiento.
reduce es una Accin en el RDD.
Acciones: collect(), count(), take(x), takeOrdered(x,y), takeSample(x, y), countByKey(), foreach(funcion())

```python
#Ejemplo
r = sc.parallelize(range(3), 1) #una particin (0, 1, 2)
r.reduce(lambda x,y: x-y)
#(0-1) - 2 
```

```python
r = sc.parallelize(range(3), 2) #Dos particiones {0} y {1, 2}
r.reduce(lambda x,y: x-y)
# (0) y (1 -2)
#  0      -1
#  0  - ( -1 )
```

```python
#Manejo de cadenas. Utilizacin del mtodo aggregate.
#Tres parametros: 1) Un valor inicial para el acumulador, zeroValue que tendr tipo C
#   2) una funcin seqop para combinar elementos de nuestro RDD (de tipo T)
#  con el acumulador de tipo C, devolviendo un valor de tipo C (C x T -> C).
#   3) Una funcin combOp para combinar dos acumuladores de tipo C y devolver un valor de tipo C.
#Cuenta las h en el RDD
r = sc.parallelize(["hola", "hi", "ciao"])
#          (int, cuenta las "h", suma los acumuladores)  => (valor_inicial, map, reduce)
#                      MAP                         Reduce
r.aggregate(0, lambda c, s : c + s.count("h"), lambda c1, c2: c1 + c2)
#(h, 1)
#(h, 1)
#(h, 0)
```

# Transformaciones


```python
#Transformacion map
#Permite aplicar una funcion elemento a elemento en un RDD
#RDD con elementos tipo T y una funcin f que acepta elementos T 
#y produce valores tipo V, la transformacin
#map(f) generar un RDD de elementos tipo V
r = sc.parallelize([1,2,3,4])
r2 = r.map(lambda x: x + 1)
r2.collect()
```

```python
#Sin lambda
def increment(x):
    return x + 1

r = sc.parallelize([1,2,3,4])
r2 = r.map(increment)
r2.collect()
```

```python
r = sc.parallelize(["hola", "hi", "ciao"])
r2 = r.map(lambda x: len(x))
r2.collect()
```

```python
import csv
r = sc.parallelize(["1,5,7","8,2,4"]) #cadenas en formato csv
r2 = r.map(lambda x: list(csv.reader([x]))[0]) #como resultado una lista/tupla. Una listra dentro de otra lista
#por eso se utiliza [0]
r2.collect()
#Nota: el resultado como se puede ver son cadenas de texto y NO NUMEROS
```
